```python
frequency_of_the_tuning_fork=261.6
```

```python
sampling_frequency=44100
```

```python
Number=2*sampling_frequency
```

```python
import numpy as np
```

```python
ts=np.linspace(0,2,Number)
```

```python
ys=np.sin(2*np.pi*frequency_of_the_tuning_fork*ts)
```

```python
from IPython.display import Audio
```

```python
Audio(ys,rate=sampling_frequency)
```
